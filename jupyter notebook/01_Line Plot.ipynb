## Import Libraries 
- Seaborn automatically installs these libraries 
- *numpy*
- *Scipy*
- *Pandas*
- *Matplotlib*

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool

#Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

## Adding Title 

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool

#Draw a line plot
plt.title("Phoolon ka Plot")
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

## Adding limit Y-Limit and X-Limit 

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool

#Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Set styles 
- whitegrid
- dark
- white
- ticks

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool

#Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show()
```

## Size of figure

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool

#Size of figure
plt.figure(figsize=(6, 4))
#Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

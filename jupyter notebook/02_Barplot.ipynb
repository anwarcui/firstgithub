## Barplot

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt

#Load dataset 
phool = sns.load_dataset("iris")
phool
```

```python
#Draw a line plot
sns.barplot(x="species", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
#Draw a line plot
sns.barplot(x="species", y="petal_length", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
# Import Libraries 
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import mean

#Load dataset 
kashti = sns.load_dataset("titanic")
kashti

#Draw a line plot
sns.barplot(x="class", y="fare", hue="who", data=kashti,  estimator=mean)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
# Import Libraries 
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

#Load dataset 
kashti = sns.load_dataset("titanic")
kashti

#Draw a line plot
sns.barplot(x="class", y="fare", hue="sex", data=kashti,  estimator=median, saturation=0.5)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
#horizental Plot
# Import Libraries 
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

#Load dataset 
kashti = sns.load_dataset("titanic")
kashti

#Draw a line plot
sns.barplot(x="fare", y="class", hue="sex", data=kashti,  estimator=median, saturation=1)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
# Import Libraries 
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

#Load dataset 
kashti = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=kashti, linewidth=1, facecolor=(1,1,1,0), errcolor=".2", edgecolor=".2")
```
